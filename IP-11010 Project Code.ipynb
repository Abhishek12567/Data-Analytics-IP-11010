How does expected salary vary based on GPA, Family Income, Experience?

**Average Salary vs Gpa**

```python
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd


df['GPA_Range'] = pd.cut(df['CGPA'],
                         bins=[0,7,8,9,10],
                         labels=['<7','7-8','8-9','9+'])

salary_gpa = df.groupby('GPA_Range')['Expected salary (Lac)'].mean()

display(salary_gpa.round(3))

plt.figure(figsize=(8,5))
plt.bar(salary_gpa.index, salary_gpa.values, edgecolor='black')

plt.ylim(salary_gpa.min()-1, salary_gpa.max()+1)
plt.title("Average Salary by GPA Range")
plt.ylabel("Expected Salary (Lac)")

plt.tight_layout()
plt.show()
```

**Which event attracts more students from specific fields?**

```python


import pandas as pd
import matplotlib.pyplot as plt


event_field = pd.crosstab(df['Events'], df['Designation'])


top_events = event_field.sum(axis=1).nlargest(5).index
top_fields = event_field.sum().nlargest(5).index

filtered = event_field.loc[top_events, top_fields]


display(filtered)


percent = filtered.div(filtered.sum(axis=1), axis=0) * 100


percent.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Top 5 Events by Top 5 Fields (%)")
plt.ylabel("Percentage")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

**High CGPA + High Experience → Higher Salary?**

```python

high_students = df[
    (df['CGPA'] > df['CGPA'].mean()) &
    (df['Experience with python (Months)'] > df['Experience with python (Months)'].mean())
]

salary_compare = pd.Series({
    'High Performers Avg Salary': high_students['Expected salary (Lac)'].mean(),
    'Overall Avg Salary': df['Expected salary (Lac)'].mean()
})

display(salary_compare.round(3))

plt.bar(salary_compare.index, salary_compare.values)
plt.title("High Performers vs Overall Salary")
plt.ylabel("Expected Salary (Lac)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
